```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import FinanceDataReader as fdr

# 2023 애플 주식 데이터 수집
df = fdr.DataReader('AAPL','2023')

# 데이터 확인
print(df.head())
print(df.tail())
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 데이터 정보확인
df.describe()
```

```python
# 주식 가격 시각화하기
plt.figure(figsize=(10, 6))
sns.lineplot(x=df.index, y=df['Close'])

plt.title('Apple Stock Price in 2023')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# 한달 간격으로 다운샘플링
df_month=df.resample("BM").mean()
print(df_month.head())
print(df_month.tail())
```

```python
# 수익률 변수 추가하기
df_month['rtn']=df_month['Close'].pct_change()
print(df_month.head())
```

```python
df_month.describe()
```

```python
df_month.info()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_month.index, y=df_month['rtn'])

plt.title('Apple Stock Returns in 2023-2024')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.show()
```

```python
# 2달 종가 이동평균 시각화
df_month['MA']=df_month['Close'].rolling(2).mean()
print(df_month.head())
```

```python
df_month.iloc[:,[3,6]].plot(figsize=(15,8))
```

```python
df_month['MA']
```

```python
last_close=df_month['MA'].iloc[-2] # 이동평균선 60일 전 종가
print(last_close)


price=df_month['Close'].iloc[-1] # 오늘 종가
print(price)

if price>last_close:
    print('상승 장')
elif price<last_close:
    print('하락 장')
else:
    print('변화없음')
```
